## Ejercicio 1

Contar cuántos valores hay en una lista de enteros. Contar solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
contador = 0
for variable_iteradora in nums: #A LA VARIABLE ITERADORA LE PUEDO DAR CUALUIER NOMBRE
  contador = contador + 1
  if variable_iteradora == 0: #LA CONDICION QUE SE LE PONE
    break
#Respuesta:
#Debe dar 5
print(contador)
```

## Ejercicio 2

Sumar los valores que hay en una lista de enteros. Sumar solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
suma = 0
#Respuesta:
for num in nums:
  suma += num
  if num == 0:
    break
  #Debe dar 21
print(suma)
```

```python
nums = [6,8,5,2,0,6,5] #ALTERNATIVA A LA SOLUCION DE ARRIBA AL EJERCICIO 2
suma = 0
for numero in nums:
  suma = suma + numero
  if numero == 0:
    break
print(suma)
```

## Ejercicio 5 (Opcional)

Crear una lista a partir de una secuencia de valores de entrada terminada en *.
Una vez creada, modificarla elevando al cuadrado todos los pares y al cubo todos los impares, y mostrarla.

```python
lista = [3,56,6,7,"*",2,78]
valor = "*"
numero = int
#Respuesta:
for numero in lista:
  if numero == valor:
    break
  if numero % 2 == 0:
    cuadrado = numero * numero
    print(cuadrado)
  if numero % 2 != 0:
    cubo = numero * numero * numero
    print(cubo)

#Debe dar [27, 3136, 36, 343]
```

```python
lista = [3,56,6,7,"*",2,78]
new_lista = []
for elemento in lista:
  if elemento == '*':
    break
  new_lista.append(elemento)
  #print(new_lista)
lista_resultado = []
for numero in new_lista:
  if numero % 2 == 0:
    lista_resultado.append(numero ** 2)
  if numero % 2 != 0:
    lista_resultado.append(numero ** 3)
print(lista_resultado)

```
