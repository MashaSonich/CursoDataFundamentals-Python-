## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
#Rellena la variable query con tu consulta
query = """  SELECT DISTINCT(LaunchSite)
             FROM SPACEXTBL  """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
#Rellena la variable query con tu consulta
query1 = """ SELECT * FROM SPACEXTBL WHERE LaunchSite LIKE 'CCS%' """
df  = pd.read_sql(query1, con = engine )
df[0:5]
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query2 = """ SELECT SUM(PayloadMass) FROM `SPACEXTBL` WHERE `Customers` LIKE 'NASA (CRS)' """
df = pd.read_sql(query2, con = engine)
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query3 = """ SELECT AVG (PayloadMass)
             FROM `SPACEXTBL`
             WHERE `Serial` LIKE '%B1%' """
df  = pd.read_sql(query3, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query4 = """ SELECT LandingOutcome, MIN(Date)
             FROM SPACEXTBL
             WHERE LandingOutcome = 'True RTLS' """
df  = pd.read_sql(query4, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query5 = """ SELECT (Serial), (LandingOutcome), (PayloadMass)
            FROM SPACEXTBL
            WHERE (PayloadMass) BETWEEN 4000 AND 6000
            AND (LandingOutcome) = 'True ASDS' """
df = pd.read_sql(query5, con = engine)
df
```

### Ejercicio 7
